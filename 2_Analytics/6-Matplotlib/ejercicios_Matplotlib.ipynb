```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 1000)

fig, ax = plt.subplots(1, 2, figsize=(5, 3))  # Ajusté el ancho de la figura
ax[0].plot(x, 3*x)
ax[0].set(xlim=(0, 5), ylim=(-15, 15))
ax[1].plot(x, -3*x-5)
ax[1].set(xlim=(0, 5), ylim=(-15, 15))

for axi in ax:
    axi.set_xlabel('Eje X')
    axi.set_ylabel('Eje Y')

ax[0].set(title='Recta sube')
ax[1].set(title='Recta baja')

# Ajustar el espacio horizontal entre los subgráficos
plt.tight_layout()

plt.show()
```
