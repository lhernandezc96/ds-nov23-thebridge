# OpenStreetMap

```python
# Amenities
# https://wiki.openstreetmap.org/wiki/Key:amenity
overpass_url = "http://overpass-api.de/api/interpreter"
overpass_query = """
[out:json];
node["amenity"="cafe"]
  (40.420801, -3.694702,40.423754, -3.688167);
out;
"""
response = requests.get(overpass_url,
                        params={'data': overpass_query})
print(response)
data = response.json()
```

```python
# Acceder a los valores del diccionario para guardarlos en variables para luego crear un Dataframe
dict_cafes = {"lat":[], "lon":[], "street":[], "number_street":[],"name":[]}
for i in data["elements"]:
    
    lat = i["lat"]
    lon = i["lon"]
    street = i["tags"].get("addr:street", "Sin Calle")
    number_street = i["tags"].get("addr:housenumber", 0)
    name = i["tags"].get("name", "Sin nombre")
    
    dict_cafes["lat"].append(lat)
    dict_cafes["lon"].append(lon)
    dict_cafes["street"].append(street)
    dict_cafes["number_street"].append(number_street)
    dict_cafes["name"].append(name)

df_cafe = pd.DataFrame(dict_cafes)
df_cafe
```

```python
import numpy as np
import pandas as pd

years = np.arange(2020, 2021)
id_movies = []
release_date = []
title = []
api_key = "5254b3e1274baac4c5a450bec75bae6f"
for i in years:
    print("year:", i )
    for j in range(1,5):
        response = requests.get(f'https://api.themoviedb.org/3/discover/movie?api_key={api_key}&page={j}&primary_release_year={i}&sort_by=popularity.desc')
        highest_revenue = response.json()
        print(highest_revenue)
        first_results = highest_revenue["results"]
        for k in first_results:
            try:
                release_date.append(k["release_date"])
                title.append(k["title"])
                id_movies.append(k["id"])
            except:
                release_date.append(np.nan)
                title.append(np.nan)
                id_movies.append(np.nan)
                continue
df_movies_id = pd.DataFrame(id_movies, columns=["id"])
df_movies_id["date"] = release_date
df_movies_id["title"] = title
```
