## Ejercicio 1
Somos una empresa que tiene varias tiendas de electrodomesticos. Necesitamos un programa para manejar las tiendas, ver las ventas que han tenido, cuántos empleados hay, etc... Para ello vamos a modelizar la tienda en una clase, que tendrá las siguientes características:
* Nombre clase: "Tienda"
* Atributos comunes:
    * Tipo: "Electrodomésticos"
    * Abierta: True
* Atributos propios de cada tienda:
    * Nombre - String
    * Dirección - String
    * Número de empleados - int
    * Ventas ultimos 3 meses - Lista de 3 numeros
* Método para calcular las ventas de todos los meses, que devuelva un numero con todas las ventas.
* Método que calcula la media de ventas de los ultimos meses, por empleado
* Método que devuelve en un string el nombre de la tienda, junto con su dirección.
* Método que obtiene las ventas del último mes.
* Método para dar una proyección de las ventas en caso de haber invertido X dinero en marketing en esa tienda. Siendo X el parámetro de entrada. Si X es menor de 1000, las ventas de los ultimos 3 meses hubiesen sido de (1.2 x ventas), si es mayor o igual de 1000, las ventas hubiesen sido de (1.5 x venta). El método reescribe el atributo de ventas en los últimos 3 meses, y además devuelve las ventas con el aumento.

Se pide:
1. Implementar la clase "Tienda"
2. Crear tres tiendas con datos inventados
3. Comprueba en al menos una de ellas, todo lo implementado en la clase tienda (sus atributos, media de ventas, ventas/empleado...)
3. Calcular las ventas del último mes de todas las tiendas. Para ello usa el bucle `for`
4. Imprime por pantalla los nombres de las tiendas cuya dirección lleve el string "Avenida"
5. Seguro que ya has hecho este apartado... Documenta la clase :)

````python
class Tienda:
    """
    Clase Tienda:
    Atributos:
    - tipo (str): Tipo de tienda (fijo: "Electrodomésticos").
    - abierta (bool): Indica si la tienda está abierta (fijo: True).
    - nombre (str): Nombre de la tienda.
    - direccion (str): Dirección de la tienda.
    - num_empleados (int): Número de empleados en la tienda.
    - ventas_ultimos_3_meses (list): Lista con las ventas de los últimos 3 meses.

    Métodos:
    - __init__(self, nombre, direccion, num_empleados, ventas_ultimos_3_meses): Constructor de la clase.
    - calcular_ventas_totales(self): Calcula y devuelve la suma de las ventas de los últimos 3 meses.
    - calcular_ventas_totales_por_empleado(self): Calcula y devuelve la media de ventas por empleado.
    - nombre_y_direccion(self): Devuelve un string con el nombre y la dirección de la tienda.
    - ventas_ultimo_mes(self): Devuelve las ventas del último mes.
    - proyeccion_ventas_marketing(self, inversion_marketing): Calcula y devuelve las ventas proyectadas con base en una inversión de marketing.
    """
    def __init__(self, nombre, direccion, num_empleados, ventas_ultimos_3_meses):
        self.tipo = "Electrodomésticos"
        self.abierta = True
        self.nombre = nombre
        self.direccion = direccion
        self.num_empleados = num_empleados
        self.ventas_ultimos_3_meses = ventas_ultimos_3_meses   

    def calcular_ventas_totales(self):
         return sum(self.ventas_ultimos_3_meses)
    def calcular_ventas_totales_por_empleado(self):
           return sum(self.ventas_ultimos_3_meses) / self.num_empleados
    def nombre_y_direccion(self):
        return self.nombre + " " + self.direccion
    def ventas_ultimo_mes(self):
        return self.ventas_ultimos_3_meses[-1]
    def proyeccion_ventas_marketing(self, inversion_marketing):
        aumento_ventas = []
        for i in range(len(self.ventas_ultimos_3_meses)):
            if inversion_marketing < 1000:
                aumento_ventas.append(self.ventas_ultimos_3_meses[i] * 1.2)
            else:
                aumento_ventas.append(self.ventas_ultimos_3_meses[i] * 1.5)

            self.ventas_ultimos_3_meses[i] = aumento_ventas[i]

        return aumento_ventas
        
````

````python
ejemploTienda1 = Tienda("Electronica Express", "Calle Electrón nº314", 4, [30000, 32000, 33000])
ejemploTienda2 = Tienda("ElectronicaNow", "Calle Protón nº314", 5, [25000, 30000, 37000])
ejemploTienda3 = Tienda("ElectroCasa", "Avenida Electrón nº314", 9, [33000, 37000, 40000])

print(ejemploTienda1.tipo, ejemploTienda1.abierta, ejemploTienda1.nombre, ejemploTienda1.direccion, ejemploTienda1.num_empleados, ejemploTienda1.ventas_ultimos_3_meses)
print(ejemploTienda1.calcular_ventas_totales(), ejemploTienda1.calcular_ventas_totales_por_empleado(), ejemploTienda1.nombre_y_direccion())
InvMarketing = 1200
print(ejemploTienda1.proyeccion_ventas_marketing(InvMarketing))
print(ejemploTienda1.ventas_ultimos_3_meses)

ventas_ultimo_mes_todas_tiendas = [tienda.ventas_ultimo_mes() for tienda in [ejemploTienda1, ejemploTienda2, ejemploTienda3]]
print("Ventas del último mes de todas las tiendas: ", ventas_ultimo_mes_todas_tiendas)

tiendas_con_avenida = [tienda for tienda in [ejemploTienda1, ejemploTienda2, ejemploTienda3] if "Avenida" in tienda.direccion]
print("Tiendas con 'Avenida' en la dirección:")
for tienda in tiendas_con_avenida:
    print(tienda.nombre)
print(Tienda.__doc__)

````

## Ejercicio 2
En este ejercicio vamos a implementar una clase *Perro* en Python. La clase tiene las siguientes características:
* Cosas que sabemos seguro que tiene un perro
    * Tiene 4 patas
    * 2 orejas
    * 2 ojos
    * Una velocidad de 0. Por defecto, el perro está parado
* Cuando se inicialice:
    * El perro será de una determinada raza
    * Por defecto tendrá pelo "Marrón", a no ser que se diga lo contrario.
    * Por defecto no tendrá dueño, a no ser que se diga lo contrario.
    
* Dispondrá también de un método llamado andar, que tiene un argumento de entrada (aumento_velocidad). Este valor se le sumará a la velocidad que ya llevaba el perro.
* Necesita otro método (parar), donde pondremos la velocidad a 0.
* Otro método llamado "ladrar", que tendrá un argumento de entrada, y la salida será el siguiente string: "GUAU!" + el argumento de entrada.


Se pide:
* Implementa la clase *Perro*
* Crea un objeto de tipo *Perro*, sin dueño
* Comprueba que están bien todos sus atributos
* Prueba que ande, y comprueba su velocidad
* Páralo
* Documenta la clase *Perro*

````python
class Perro:
    """
    Clase Perro:

    Atributos:
    - patas (int): Número de patas (fijo: 4).
    - orejas (int): Número de orejas (fijo: 2).
    - ojos (int): Número de ojos (fijo: 2).
    - velocidad (int): Velocidad actual del perro (inicialmente 0).
    - raza_perro (str): Raza del perro (especificada al crear la instancia).
    - color_perro (str): Color del pelaje del perro (predeterminado: "Marrón").
    - dueño_perro (bool): Indica si el perro tiene un dueño (predeterminado: False).

    Métodos:
    - __init__(self, raza_perro, color_perro="Marrón", dueño_perro=False):
      Constructor de la clase. Inicializa los atributos al crear una instancia del perro.
    
    - andar(self, aumento_velocidad):
      Aumenta la velocidad del perro en la cantidad especificada y devuelve la nueva velocidad.

    - parar(self):
      Detiene al perro, estableciendo su velocidad a 0.

    - ladrar(self, ladrido):
      Produce un ladrido del perro, agregando el ladrido dado al sonido "GUAU!".

    Ejemplo de uso:
    ```python
    # Crear una instancia de la clase Perro
    mi_perro = Perro(raza_perro="Labrador", color_perro="Negro", dueño_perro=True)

    # Hacer que el perro ande y ladre
    mi_perro.andar(5)
    print(f"Velocidad actual del perro: {mi_perro.velocidad}")
    
    ladrido = mi_perro.ladrar("¡Hola!")
    print(f"Sonido de ladrido: {ladrido}")

    # Detener al perro
    mi_perro.parar()
    print(f"Velocidad después de detener al perro: {mi_perro.velocidad}")
    ```
    """
    def __init__(self, raza_perro, color_perro="Marrón", dueño_perro=False):
        """
        Constructor de la clase Perro.

        Parámetros:
        - raza_perro (str): Raza del perro.
        - color_perro (str): Color del pelaje del perro (predeterminado: "Marrón").
        - dueño_perro (bool): Indica si el perro tiene un dueño (predeterminado: False).
        """
        self.patas = 4
        self.orejas = 2
        self.ojos = 2
        self.velocidad = 0
        self.raza_perro = raza_perro
        self.color_perro = color_perro
        self.dueño_perro = dueño_perro
    
    def andar(self, aumento_velocidad):
        """
        Aumenta la velocidad del perro en la cantidad especificada.

        Parámetros:
        - aumento_velocidad (int): Cantidad en la que se incrementa la velocidad.

        Retorna:
        - int: Nueva velocidad del perro.
        """
        self.velocidad += aumento_velocidad
        return self.velocidad

    def parar(self):
        """Detiene al perro, estableciendo su velocidad a 0."""
        self.velocidad = 0

    def ladrar(self, ladrido):
        """
        Produce un ladrido del perro, agregando el ladrido dado al sonido "GUAU!".

        Parámetros:
        - ladrido (str): Sonido específico del ladrido.

        Retorna:
        - str: Sonido completo del ladrido del perro.
        """
        return "GUAU! " + ladrido
````
